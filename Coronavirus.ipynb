### 1. Getting started with Data exploration and cleaning steps

#### 1.1 Understanding the columns of the data

- status_id ------------ unique id for each tweet
- user_id -------------- id of the user
- created_at ----------- timestamp of the tweet
- screen_name ---------- name of the user on twitter
- text ----------------- content of the tweet
- source --------------- source of the tweet (like android, iphone etc..)
- reply_to_status_id --- status_id of the tweet to which the current user replied (Null if tweet is not a reply)
- reply_to_user_id ----- user_id of the tweet to which current user replied (Null if tweet is not a reply)
- reply_to_screen_name - screen_name of the tweet to which current user replied (Null if tweet is not a reply)
- is_quote ------------- TRUE if tweet is a quote otherwise FALSE
- is_retweet ----------- TRUE if tweet is a retweet otherwise FALSE
- favourites_count ----- count of users who bookmarked the tweet
- retweet_count -------- count of retweets for the tweet
- country_code --------- code of the user's Country
- place_full_name ------ place of the name user
- place_type ----------- type of place (like city etc..)
- followers_count ------ count of followers of the user
- friends_count -------- count of users, current user is following
- account_lang --------- all the values are NULL
- account_created_at --- date of joining in twitter for the user
- verified ------------- TRUE if the account belongs to verified user otherwise FALSE
- lang ----------------- language of the tweet

#### 1.3 Loading Dataset for 15 days from 16-04-202 to 30-04-2020

```python
%%time

for x in range(16,31):   
    a = pd.read_csv('2020-04-{} Coronavirus Tweets.csv'.format(x))
    
    #creating column for date
    a['date'] = a['created_at'].apply(lambda x : dateutil.parser.isoparse(x).date())

    #creating column for hour
    a['hour'] = a['created_at'].apply(lambda x : dateutil.parser.isoparse(x).hour)

    #Dropping unnecessary columns
    a.drop(['created_at','screen_name','reply_to_status_id',
           'reply_to_screen_name','place_full_name','place_type','followers_count',
           'friends_count','account_lang','account_created_at'], axis = 1,inplace=True)

    #Filtering rows with language as 'English' and having non Null Country code
    a = a[(a['lang']=='en') & (a['country_code']).notnull()] .reset_index(drop=True)

    #creating a new column list of "Hashtag"
    a['hashtag'] = a['text'].apply(lambda x: re.findall(r'\B#\w*[a-zA-Z]+\w*', x))

    #creating a new column list of "Mention"
    a['mention'] = a['text'].apply(lambda x: re.findall(r'\B@\w*[a-zA-Z]+\w*', x)) 

    #creating a new column list of "Hyperlink"
    a['hyperlink'] = a['text'].apply(lambda x: re.findall(r'\b((?:https?://)?(?:(?:www\.)?(?:[\da-z\.-]+)\.(?:[a-z]{2,6})|(?:(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.){3}(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)|(?:(?:[0-9a-fA-F]{1,4}:){7,7}[0-9a-fA-F]{1,4}|(?:[0-9a-fA-F]{1,4}:){1,7}:|(?:[0-9a-fA-F]{1,4}:){1,6}:[0-9a-fA-F]{1,4}|(?:[0-9a-fA-F]{1,4}:){1,5}(?::[0-9a-fA-F]{1,4}){1,2}|(?:[0-9a-fA-F]{1,4}:){1,4}(?::[0-9a-fA-F]{1,4}){1,3}|(?:[0-9a-fA-F]{1,4}:){1,3}(?::[0-9a-fA-F]{1,4}){1,4}|(?:[0-9a-fA-F]{1,4}:){1,2}(?::[0-9a-fA-F]{1,4}){1,5}|[0-9a-fA-F]{1,4}:(?:(?::[0-9a-fA-F]{1,4}){1,6})|:(?:(?::[0-9a-fA-F]{1,4}){1,7}|:)|fe80:(?::[0-9a-fA-F]{0,4}){0,4}%[0-9a-zA-Z]{1,}|::(?:ffff(?::0{1,4}){0,1}:){0,1}(?:(?:25[0-5]|(?:2[0-4]|1{0,1}[0-9]){0,1}[0-9])\.){3,3}(?:25[0-5]|(?:2[0-4]|1{0,1}[0-9]){0,1}[0-9])|(?:[0-9a-fA-F]{1,4}:){1,4}:(?:(?:25[0-5]|(?:2[0-4]|1{0,1}[0-9]){0,1}[0-9])\.){3,3}(?:25[0-5]|(?:2[0-4]|1{0,1}[0-9]){0,1}[0-9])))(?::[0-9]{1,4}|[1-5][0-9]{4}|6[0-4][0-9]{3}|65[0-4][0-9]{2}|655[0-2][0-9]|6553[0-5])?(?:/[\w\.-]*)*/?)\b', x))
    
    print('2020-04-{} Coronavirus Tweets.csv ---- appended'.format(x))
    
    if x==16:
        data = a.copy()
    else:
        data = data.append(a, ignore_index=True, sort=False)
```

### 3. Text Pre-processing

- Tokenization
- Punctuation removal
- special character treatment
- POS tagging
- Lemmatization
- Stopwords removal
- Removing other unnecessary words

```python
%%time
# Sentence Tokenization
data.loc[:,'sentok'] = data['text'].apply(sent_tokenize) 
sentlen = data.sentok.apply(len)

print('Avg No of sentences in a review : ' + str(np.mean(sentlen)))
print('Min No of sentences in a review : ' + str(np.min(sentlen)))
print('Max No of sentences in a review : ' + str(np.max(sentlen)))
```

```python
%%time
# Word Tokenization
data.loc[:,'wordtok'] = data['text'].apply(word_tokenize)
wrdcnt = data.wordtok.apply(len)

print('Avg No of words in a review : ' + str(np.mean(wrdcnt)))
print('Min No of words in a review : ' + str(np.min(wrdcnt)))
print('Max No of words in a review : ' + str(np.max(wrdcnt)))
```

```python
def punctuation_trimming(wd):
    y = ''.join((x for x in wd if x not in string.punctuation))
    return y

def special_char_removal(wd):
    y = re.sub('[^A-Za-z0-9]+', ' ', wd).strip()
    return y

#Lemmatizing nouns, verbs, adjectives, adverbs as others don't alter much after lemmatizing
def nltk_tag_to_wordnet_tag(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:          
        return None

def lemmatize(wd):
    nltk_tagged = nltk.pos_tag([wd])  
    tag = nltk_tag_to_wordnet_tag(nltk_tagged[0][1])
    if tag is None:
        return wd
    else:
        return WordNetLemmatizer().lemmatize(wd, tag)

def remove_extra_words(wd):
    y = [x if x not in ['covi d19','covid'] else 'covid19' for x in wd]
    y = [x for x in y if not len(x)<3 and x not in ('http','amp') and x.isdigit() == False]
    return y

def clean_list(wd):
    new_list = [x.lower() for x in wd]
    new_list = [punctuation_trimming(x) for x in new_list]
    new_list = [special_char_removal(x) for x in new_list]
    new_list = [x for x in new_list if x!='']
    new_list = [lemmatize(x) for x in new_list]
    new_list = [x for x in new_list if x not in set(stopwords.words('english'))]
    new_list = remove_extra_words(new_list)
    return new_list
```

```python
%%time
#creating new column for cleaned tokens
data['wordtok2'] = data['wordtok'].apply(lambda x : clean_list(x))
```

### 4. Topic Modelling

- Building topics from the tweets using gensim

```python
%%time
#Creating dictionary of bag of words
dictionary = gensim.corpora.Dictionary(data['wordtok2'])
#filtering out words that appeared in less than 15 tweets, more than 80% of tweets and considering top 100000 words
dictionary.filter_extremes(no_below=15, no_above=0.8, keep_n=100000)

#freq matrix of words
bow_corpus = [dictionary.doc2bow(doc) for doc in data['wordtok2']]

#Building TF-IDF matrix
tfidf = models.TfidfModel(bow_corpus)
corpus_tfidf = tfidf[bow_corpus]
```

```python
%%time
#Creating 10 different topics using LDA
lda_model_tfidf = gensim.models.LdaMulticore(corpus_tfidf, num_topics=10, id2word=dictionary, passes=2, workers=4)

#Display characteristics of each topic
lda_model_tfidf.print_topics(num_words=15)
```

```python
%%time
#Assigning Topic to all tweets
topic = []
for x in bow_corpus:
    a = lda_model_tfidf[x]
    b = [a[y][1] for y in range(len(a))]
    topic.append(np.argmax(b)+1)
    
data = pd.concat([data, pd.DataFrame(topic,columns=['topic'])], axis=1)
```

### 5. Sentiment Analysis

Categorizing the Tweets based on their polarity
- Postive  (if polarity > 0.05)
- Neutral  (if polarity > -0.05 and <= 0.05)
- Negative (if polarity <= -0.05)

```python
%%time
#### Calculating score for each tweet in the dataframe using vader sentiment analysis
analyser = SentimentIntensityAnalyzer()
data['polarity'] = data['text'].apply(lambda x : analyser.polarity_scores(x)['compound'])
data['Sentiment'] = data['polarity'].apply(lambda x : 'Positive' if x>0.05 else 'Neutral' if x>-0.05 else 'Negative')
```
